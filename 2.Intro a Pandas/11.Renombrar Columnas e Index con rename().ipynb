```python
import pandas as pd
```

```python
# leer los archivos csv
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# Renombrar Columna

```python
# renombrar una columna y sobreescribir los valores del dataframe
df_exams = df_exams.rename(columns={'gender':'Gender'})
```

```python
# renombrar una columna y actualizar el dataframe con el argumento inplace
df_exams.rename(columns={'math score':'MS', 'reading score':'RS', 'writing score':'WS'}, inplace=True)
```

```python
# mostrar dataframe
df_exams.head()
```

# Renombrar Index

```python
# renombrar index 0, 1 y 2 y actualizar el dataframe
df_exams.rename(index={0:'A', 1:'B', 2:'C'}, inplace=True)
```
