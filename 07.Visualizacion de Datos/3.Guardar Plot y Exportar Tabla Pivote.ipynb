```python
import pandas as pd
```

```python
# leer el archivo csv
df_population_raw = pd.read_csv('population_total.csv')
```

```python
# mostrar el dataframe
df_population_raw
```

# Hacer una Tabla Pivote

.pivot(): Devuelve un dataframe reformado que esta organizado por cierto index/valores de columnas ("tabla pivote sin funciones agregadas) 

```python
# eliminar valores nulos
df_population_raw.dropna(inplace=True)
```

```python
# hacer tabla pivote
df_pivot = df_population_raw.pivot(index='year', columns='country',
                                        values='population')
```

```python
# seleccionar algunos paises
df_pivot = df_pivot[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
```

```python
# mostrar tabla pivote
df_pivot
```

```python
# exportar tabla pivote
df_pivot.to_excel('pivot_table.xlsx')
```

```python
# importar matplotlib
import matplotlib.pyplot as plt
```

# Lineplot (Grafico de linea)

```python
df_pivot.plot(kind='line',
              xlabel='Years', ylabel='Population',
              title='Population (1955-2020)',
              figsize=(8,4))

# guardar figure
plt.savefig('mi_figura.png')
# mostrar plot
plt.show()
```
