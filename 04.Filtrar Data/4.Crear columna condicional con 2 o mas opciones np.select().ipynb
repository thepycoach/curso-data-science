# Crear una columna condicional con 2 o mas opciones: np.select()

```python
import numpy as np
```

```python
# crear un array basado en multiple niveles de precio (+2 opciones)

# crear "conditions" (condiciones) y "values" (valores)
conditions = [
    (df_laptops['Price_euros']>3000),
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    (df_laptops['Price_euros']<=800),
]
choices = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# aÃ±adir a nueva columna
df_laptops['Price_tier'] = np.select(conditions, choices)
```

```python
# mostrar dataframe
df_laptops
```

```python
# contar los valores en columna price_tier 
df_laptops.value_counts('Price_tier')
```

```python
# crear un array basado en multiple tamaÃ±os de pantalle ("screen size" +2 opciones)
# Muy grande>16, Grande>14, PequeÃ±o>12, Muy PequeÃ±o<12

# crear "conditions" (condiciones) y "values" (valores)
conditions = [
    (df_laptops['Inches']>16),
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    (df_laptops['Inches']<=12),
]
choices = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# aÃ±adir a nueva columna
df_laptops['Screen_size'] = np.select(conditions, choices)
```

```python
# mostrar dataframe
df_laptops
```

```python
# contar los valores en columna Screen_size
df_laptops.value_counts('Screen_size')
```
