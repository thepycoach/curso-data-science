```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

# Filtrar dataframe basado en una condicion

```python
# encontrar filas que tengan "Apple" en la columna "Company"
df_laptops['Company']=='Apple'
```

```python
# filtrar dataframe basado en la condicion
df_laptops[df_laptops['Company']=='Apple']
```

```python
# verificar que compañias estan el la columna "Company" con el metodo value_counts()
df_laptops[df_laptops['Company']=='Apple'].value_counts('Company')
```
