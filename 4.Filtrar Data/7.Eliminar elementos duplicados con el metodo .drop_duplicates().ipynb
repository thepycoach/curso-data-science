```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

# drop_duplicates()

```python
# eliminar duplicados en 2 o mas columnas
df_laptops.drop_duplicates(['Company'])
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# ordenar dataframe ascendiente por compa√±ia (company) y precio (price)
# mas barato primero y mas caro ultimo
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# mas barato: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# mas caro: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```

```python
# Argumentos

# inplace: eliminar duplicados en el dataframe o devuelve copia
# ignore_index=True: el index resultado tendra etiqueta 0, 1, ‚Ä¶
```
