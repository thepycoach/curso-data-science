```python
import pandas as pd
```

```python
# leer archivo csv
df = pd.read_csv('players_20.csv')
# fijar index
df.set_index('short_name', inplace=True)
# seleccionar columnas
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# mostrar dataframe
df
```

# Seleccionar con un valor

loc[nombre_fila, nombre_columna]

```python
# obtener toda la data sobre L.Messi
df.loc['L. Messi']
```

```python
# obtener el tamaño de L.Messi
df.loc['L. Messi', 'height_cm']
```

```python
# obtener el peso de Cristiano Ronaldo
df.loc['Cristiano Ronaldo', 'weight_kg']
```

```python
# obtener todas las filas dentro de columna 'height_cm'
df.loc[:, 'height_cm']
```

```python
# obtener todas las columnas que corresponder al index 'L. Messi'
df.loc['L. Messi', :]
```

# Seleccionar con una lista de valores

```python
# obtener data sobre L.Messi y Cristiano Ronaldo
df.loc[['L. Messi', 'Cristiano Ronaldo']]
```

```python
# obtener el tamaño de L.Messi y Cristiano Ronaldo
df.loc[['L. Messi', 'Cristiano Ronaldo'], 'height_cm']
```

```python
# obtener el tamaño y peso de L.Messi
df.loc['L. Messi', ['height_cm', 'weight_kg']]
```

```python
# obtener el tamaño y peso de L.Messi y Cristiano Ronaldo
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm', 'weight_kg']]
```

# Seleccionar un rango de data con un slice (rebanada)

start:stop:step  (Nota: Contrariamente a las rebanadas tipicas en Python, "start" y "stop" en este caso son incluidos)

```python
# hacer slice a nombre de columnas
players = ['L. Messi', 'Cristiano Ronaldo']

df.loc[players, 'age':'club']
```

```python
# hacer slice a los nombre de los index
columns = ['age', 'dob', 'height_cm', 'weight_kg']

# obtener los nombres de jugadores top1 y top10 
# df.index[:10]

df.loc['L. Messi':'M. Salah', columns]
```

# Seleccionar con condiciones

```python
# una condicion: seleccionar jugadores con tamaño arriba de 180cm
columns = ['age', 'dob', 'height_cm', 'weight_kg']

df.loc[df['height_cm']>180, columns]
```

```python
# multiples condiciones: seleccionar jugadores con tamaño arriba de 180cm y que son de Argentina
df.loc[(df['height_cm']>180) & (df['nationality']=='Argentina'), :]
```
